```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('annual_mean_temp')
data.head()
```

```python
%matplotlib inline
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(data.Year, data.mean_temp, color='tab:red', label='annual_mean_temp')

# Same as above
ax.set_xlabel('Year')
ax.set_ylabel('mean_temp')
ax.set_title('annual_mean_temp_evolution')
#ax.grid(True)
ax.legend(loc='upper left');
```

```python
co2 = pd.read_csv('co2_em.csv')
co2.head()
```

```python
%matplotlib inline
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(co2.Year, co2.co2_em, color='tab:red', label='Co2_evolution')

# Same as above
ax.set_xlabel('Year')
ax.set_ylabel('co2_em')
ax.set_title('Co2_evolution')
#ax.grid(True)
ax.legend(loc='upper left');
```

```python
all_data = data.merge(co2)
all_data.isna().sum()
```

```python
all_data = all_data.dropna().set_index('Year')
all_data.head(2)
```

```python
all_data.to_csv('all_data.csv')
```

```python
%matplotlib inline
plt.rc('font', size=10)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(all_data.co2_em, all_data.mean_temp, color='tab:red', label='Co2_temp_evolution')

# Same as above
ax.set_xlabel('co2_em')
ax.set_ylabel('mean_temp')
ax.set_title('Co2_temp_evolution')
#ax.grid(True)
ax.legend(loc='upper left');
```

```python
# we can see from the plot that often when the co2_emission increase, the annual mean temperature increses as well
# except in some years where even the emissions of co2 increases, the temp decreases
```
